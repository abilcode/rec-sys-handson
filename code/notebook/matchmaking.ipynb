```python
import numpy as np
import random
import pandas as pd
```

```python
def generate_interest(num):
    # create a list of user ids
    user_ids = list(range(1, num+1))

    # create a dictionary to hold binary data for each variable
    data_dict = {}
    for i in range(7):
        data_dict[f"interest_{i+1}"] = np.random.randint(0, 2, num)

    # create a dataframe from the user ids and binary data dictionary
    df = pd.DataFrame({"id": user_ids, **data_dict})
    return df

def cosine_similarity(user, mentor):
    vec1 = user
    vec2 = mentor
    cosine = np.dot(vec1,vec2)/((np.dot(vec1,vec1)*np.dot(vec2,vec2))**0.5) 
    return cosine
```

```python
# set the number of users
num_users = 100
num_mentor = 5

df_user = generate_interest(num_users)
df_mentor = generate_interest(num_mentor)
```

```python
menteeId = [i for i in df_user.id]
mentorId = [i for i in df_mentor.id]
```

```python
mentee2mentor = {}
sim_mentee={}
con = []
var = [f"interest_{i+1}" for i in range(7)]
for u in menteeId:
    mentee2mentor[u] = mentorId
    user = df_user.loc[df_user.id==u,var].values.reshape(1,-1)
    user = np.squeeze(np.asarray(user))
    print(f'user {u}',user)
    for m in mentorId:
        mentor = df_mentor.loc[df_mentor.id==m,var].values.reshape(1,-1)
        mentor = np.squeeze(np.asarray(mentor))
        print(f'mentor {m}',mentor)
        sim = cosine_similarity(user, mentor)
        con.append(np.abs(sim))
    sim_mentee[u] = con
    con = []
        
        
```

```python
mentee2mentor
df = pd.DataFrame([(k, v) for k, values in mentee2mentor.items() for v in values], columns=['user_id', 'mentor_id'])
```

```python
df['similarity'] = df['user_id'].copy()
```

```python
for row in range(df.shape[0]):
    #print(row)
    user = df.iloc[row,0]
    mentor = df.iloc[row,1]
    print(user,mentor)
    df['similarity'][row] = sim_mentee[user][mentor-1]
```

```python
df_sorted_rating = df.sort_values(by=['user_id','similarity'],ascending=[True,False]).copy()
df_sorted_rating
```
