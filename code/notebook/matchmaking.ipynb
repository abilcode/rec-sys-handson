```python
import numpy as np
import random
random.seed(42)
import pandas as pd
```

```python
def generate_interest(num):
    # create a list of user ids
    user_ids = list(range(1, num+1))

    # create a dictionary to hold binary data for each variable
    data_dict = {}
    for i in range(7):
        data_dict[f"interest_{i+1}"] = np.random.randint(0, 2, num)

    # create a dataframe from the user ids and binary data dictionary
    df = pd.DataFrame({"id": user_ids, **data_dict})
    return df

def cosine_similarity(user, mentor):
    vec1 = user
    vec2 = mentor
    cosine = np.dot(vec1,vec2)/((np.dot(vec1,vec1)*np.dot(vec2,vec2))**0.5) 
    return cosine
```

```python
# set the number of users
num_users = 100
num_mentor = 10

df_user = generate_interest(num_users)
df_mentor = generate_interest(num_mentor)
```

```python
menteeId = [i for i in df_user.id]
mentorId = [i for i in df_mentor.id]
```

```python
mentee2mentor = {}
sim_mentee={}
con = []
var = [f"interest_{i+1}" for i in range(7)]
for u in menteeId:
    mentee2mentor[u] = mentorId
    user = df_user.loc[df_user.id==u,var].values.reshape(1,-1)
    user = np.squeeze(np.asarray(user))
    #print(f'user {u}',user)
    for m in mentorId:
        mentor = df_mentor.loc[df_mentor.id==m,var].values.reshape(1,-1)
        mentor = np.squeeze(np.asarray(mentor))
        #print(f'mentor {m}',mentor)
        sim = cosine_similarity(user, mentor)
        con.append(np.abs(sim))
    sim_mentee[u] = con
    con = []
        
        
```

```python
mentee2mentor
df = pd.DataFrame([(k, v) for k, values in mentee2mentor.items() for v in values], columns=['user_id', 'mentor_id'])
```

```python
df['similarity'] = df['user_id'].copy()
```

```python
for row in range(df.shape[0]):
    #print(row)
    user = df.iloc[row,0]
    mentor = df.iloc[row,1]
    #print(user,mentor)
    df['similarity'][row] = sim_mentee[user][mentor-1]
```

```python
df_ = df.copy()
df_['rating'] = df_['similarity'].copy()
df_['rating'] = df_.rating.apply(lambda row: np.random.randint(1, 6))
df_sorted_rating = df_.sort_values(by=['user_id','similarity','rating'],ascending=[True,False,False]).copy()
df_sorted_rating.fillna(0,inplace=True)
```

```python

def generate_time_availability(n_users,n_days=7,n_hours=24,freq_h='1H') :
    # create a list of dates and hours
    dates = pd.date_range('2023-05-09', periods=n_days, freq='D')
    hours = pd.date_range('00:00', '23:00', freq=freq_h).time

    # create an empty DataFrame to store the availability schedules
    df = pd.DataFrame(columns=['User ID', 'Date', 'Start Hour','End Hour'])

    # generate random availability values for each user id and time slot
    for user_id in range(1, n_users+1):
        for date in dates:
            for hour in hours:
                availability = np.random.choice([0, 1], p=[0.4, 0.6])
                if availability :
                    df = df.append({'User ID': user_id, 'Date': date, 'Start Hour': hour}, ignore_index=True)
    
    df['End Hour'] = pd.to_timedelta(df['Start Hour'].astype(str)) + pd.to_timedelta('01:00:00')

    
    return df
```

```python
import pandas as pd
import numpy as np

# set the number of user ids, days, and hours
mentee_availability = generate_time_availability(n_users=100)
mentor_availability = generate_time_availability(n_users=10,freq_h='2H')

```
